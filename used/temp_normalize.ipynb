```python
import pandas as pd
```

```python
columns_to_encode = ['flat_toilet', 'flat_balcony', 'flat_balcony_g', 'flat_door', 'inet_type', 'flat_parking', 'live_furniture', 'flat_flooring']
df = pd.get_dummies(combined_df, columns=columns_to_encode, drop_first=True)
df
```

```python
df = df.select_dtypes(exclude='object')
df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numeric_columns = ['quadrature', 'year', 'ceiling']  # add or remove columns as needed
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have your DataFrame named df

# Get the list of numerical columns
numerical_columns =  combined_df.select_dtypes(exclude='object').columns

# For each column, drop the outliers
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter the dataframe
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Once outliers are removed, visualize your data
for x_column in numerical_columns:
    if x_column != 'price':
        plt.figure(figsize=(4, 3))
        plt.scatter(df[x_column], df['price'])
        plt.xlabel(x_column)
        plt.ylabel('Y')
        plt.title(f'Scatter Plot of {x_column} vs. Price')
        plt.show()
```

```python
# Assuming you have a DataFrame called 'df'
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(20, 15))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Set plot title
plt.title("Correlation Matrix Heatmap")

# Show the heatmap
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score

# Initialize the model with a specified number of neighbors, e.g., 5
# for i in range(3,100,2):
    # best n_neighbors is 13
knn = KNeighborsRegressor(n_neighbors=13)
knn.fit(X_train, y_train)
    
y_pred = knn.predict(X_test)
    
mae = mean_absolute_error(y_test, y_pred)
# mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
# accuracy = accuracy_score(y_test, y_pred)

print(13)
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
print()
# Predict for the new data point
y_pred_knn = knn.predict(X_new)
print(f"KNeighborsRegressor Prediction: {y_pred_knn[0]:.2f}")
print()
```

```python
import random

def generate_one_hot(prefix, options):
    if isinstance(options, list):
        if all(isinstance(x, int) for x in options):
            # if the options are all integers, just return a random choice from the list
            return random.choice(options)
        chosen_option = random.choice(options)
        return [1 if opt == chosen_option else 0 for opt in options]
    else:
        return round(random.uniform(0, 1), 4)  # generates a random float between 0 and 1

# Mapping of prefixes to their respective options
options_mapping = {
    'room_count': [1,2,3,4,5],
    'quadrature': 'continuous',
    'floor': 'continuous',
    'region': 'continuous',
    'year': 'continuous',
    'ceiling': 'continuous',
    'flat_toilet': ['no_info', 'нет', 'раздельный', 'совмещенный'],
    'flat_balcony': ['балкон', 'балкон и лоджия', 'лоджия', 'несколько балконов или лоджий'],
    'flat_balcony_g': ['да', 'нет'],
    'flat_door': ['бронированная', 'деревянная', 'металлическая'],
    'inet_type': ['no_info', 'оптика', 'проводной', 'через TV кабель'],
    'flat_parking': ['гараж', 'паркинг', 'рядом охраняемая стоянка'],
    'live_furniture': ['без мебели', 'полностью', 'частично'],
    'flat_flooring': ['дерево', 'ковролан', 'ламинат', 'линолеум', 'паркет', 'плитка', 'пробковый']
}

random_values = []

for prefix, options in options_mapping.items():
    value = generate_one_hot(prefix, options)
    if isinstance(value, list):
        random_values.extend(value)
    else:
        random_values.append(value)

print('X_new = [[', ', '.join(map(str, random_values)), ']]', sep='')
```
